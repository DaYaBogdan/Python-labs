Функции в программировании похожи на функции в математике, хотя имеют и свою специфику.
Давайте напишем какую-нибудь функцию. Для примера рассмотрим факториал. Напомним, что
факториал от натурального числа n это произведение всех натуральных чисел (начиная с 1) до . 

$factorial(n) = 1 × 2 × … × n$

Давайте для начала напишем программу для вычисления факториала какого-нибудь числа. Она будет
выглядеть так:

```python
n = 5 # присваивание переменной значения 5
f = 1 # присваивание переменной значения 5
for i in range(2, n+1): # цикл от 2 до n + 1 раз, т.е n - 1 раз
 f = f * i # умножение f на индекс цикла и присваивание f
 # эквивалентный синтаксис: f *= i
print(f) # вывод f
```

В данном случае нам нужно написать функцию, вычисляющую факториал. Она выглядит так:

```python
def factorial(n): # объявление функции факториала с параметром n
    f = 1 # присваивание в переменную 1
    for i in range(2, n+1): # цикл n - 1 раз
        f = f * i # умножение f на индекс цикла и присваивание f
    return f # возврат f
```

Теперь мы можем вызывать функцию factorial , передавая ей параметр.

```python
factorial(6) # вызов функции с передачей в качестве параметра 6
```

И даже использовать её в более сложных выражениях:

```python
factorial(5)+factorial(6) # сложение двух вызовов функций
```

Посмотрим более внимательно на то, что происходит, когда Python вычисляет значение выражения
factorial(6) . В первую очередь он смотрит на первую строчку определения функции (это так
называемая сигнатура):


def factorial(n):


Здесь он видит, что функция factorial() имеет аргумент, который называется n . Python помнит,
что мы вызвали factorial(6) , то есть значение аргумента должно быть равно 6. Таким образом,
дальше он выполняет строчку (которую мы не писали)

n = 6

После чего выполняет остальные строчки из тела функции:

f = 1
for i in range(2, n+1):
 f = f * i


Наконец он доходит до строчки

return f

В этот момент переменная f имеет значение 24. Слово return означает, что Python должен
вернуться к строчке, в которой был вызов factorial(6) , и заменить там factorial(6) на 24 (то,
что написано после return ). На этом вызов функции завершён.

__Локальные и глобальные переменные__

Внутри функции могут создаваться и использоваться различные переменные. Чтобы это не создавало
проблем, переменные, определенные внутри функции, не видны извне. Давайте рассмотрим пример:


```python
f = 10 # объявление переменной f со значением 10
def factorial(n): # функция факториала
 f = 1 # внутренняя f
 for i in range(2, n+1): # цикл n - 1 раз
    f = f * i # f * i помещается в f
 print("In the function, f =", f) # вывод
 return f # возврат f

f = 10 # объявление переменной f со значением 10 
print(f) # вывод f
print(factorial(8)) # вывод возврата функции факториала
print("Out of function") # вывод выхода из функции
print(f) # вывод f
```

Тот же код, запущенный в визуализаторе:

```python
%%tutor lang='python3' # инициализация пакета
f = 10 # объявление переменной f со значением 10
def factorial(n): # функция факториала
 f = 1 # внутренняя f
 for i in range(2, n+1): # цикл n - 1 раз
    f = f * i # f * i помещается в f
 print("In the function, f =", f) # вывод
 return f # возврат f

f = 10 # объявление переменной f со значением 10 
print(f) # вывод f
print(factorial(8)) # вывод возврата функции факториала
print("Out of function") # вывод выхода из функции
print(f) # вывод f
```

Как видно из результата выполнения этого кода, переменная f в основной программе и переменная
f внутри функции — это совсем разные переменные (визуализатор pythontutor рисует их в разных
фреймах). От того, что мы как-то меняем f внутри функции, значение переменной f вне её не
поменялось, и наоборот. Это очень удобно: если бы функция меняла значение
«внешней» переменной, то она могла бы сделать это случайно и это привело бы к непредсказуемым
последствиям.
Допустим, мы хотим написать функцию, которая будет приветствовать пользователя, используя язык,

Проблема в том, что функций, которым нужно знать, какой язык выбран, может быть очень много, и
каждый раз передавать им вручную значение переменной lang отдельным параметром довольно
мучительно. Оказывается, можно этого избежать:

```python
def hello_i18n(name, lang): # объявление функции
 if lang == 'ru': # условие выбора языка
    print("Привет,",name) # вывод имени на русском
 else: # иначе
    print("Hello,", name) # вывод имени на английском

lang = 'ru' # переменная русского языка
print("Hello world") # вывод
hello_i18n('Ivan') # запуск функции без параметра языка (вывод на русском)
lang = 'en' # переменная английского языка
hello_i18n('John') # запуск функции без параметра языка (вывод на английском)
```

Как видите, сейчс поведение функции зависит от того, чему равняется переменная lang ,
определенная вне функции. Может быть и в функции factorial() можно было обратиться к
переменной f до того момента, как мы положили в неё число 1? Давайте попробуем:


```python
def factorial(n): # объявление фукнции 
 print("In the function, before assignment, f =", f) # непроинициализирована f
 f = 1 # инициализация f
 for i in range(2, n+1): # цикл n - 1 раз
    f = f * i # присваивание в f значения f * i
 print("In the function, f =", f) # вывод
 return f # возврат значения f
```

```python
factorial(2) # вызов факториала (с ошибкой)
```

В этом случае Python выдаёт ошибку: локальная переменная f использовалась до присвоения
значения. В чём разница между этим кодом и предыдущим?

Python прежде, чем выполнить функцию, он анализирует её код и определяет, какая из переменных
является локальной, а какая глобальной. В качестве глобальных переменных по умолчанию
используются те, которые не меняются в теле функции (то есть такие, к которым не применяются
операторы типа приравнивания или += ). Иными словами, по умолчанию глобальные переменные
доступны только для чтения, но не для модификации изнутри функции.

Ситуация, при которой функция модифицирует глобальную переменную, обычно не очень
желательна: функции должны быть изолированы от кода, который из запускает, иначе вы быстро
перестанете понимать, что делает ваша программа. Тем не менее, иногда модификация глобальных
переменных необходима. Например, мы хотим написать функцию, которая будет устанавливать
значение языка пользовавателя. Она может выглядеть примерно так:

```python
def set_lang(): # вызов функции
 useRussian = input("Would you like to speak Russian (Y/N): ") # ввод языка
 if useRussian == 'Y': # используем русский?
    lang = 'ru' # значение языка русский
 else: # иначе
    lang = 'en' # английский
```

```python
lang = 'en' # значение языка англ
print(lang) # вывод переменной
set_lang() # вызов функции
print(lang) # вывод переменной
```

Как видим, эта функция не работает — собственно, и не должна. Чтобы функция set_lang смогла
менять значение переменной lang , её необходимо явно объявить как глобальную с помощью
ключевого слова global .

```python
def set_lang(): # объявление функции
 global lang # использование глобальной переменной языка
 useRussian = input("Would you like to speak Russian (Y/N): ") # ввод языка
 if useRussian == 'Y': # используем русский?
    lang = 'ru' # значение языка русский
 else: # иначе
    lang = 'en' # английский
```

```python
lang = 'en' # значение языка англ
print(lang) # вывод переменной
set_lang() # вызов функции
print(lang) # вывод переменной
```

Теперь всё работает!

## Передача аргументов
Есть разные способы передавать аргументы функции. С одним из них мы уже знакомы:

```python
def hello(name, title): # объявление функции
 print("Hello", title, name) # вывод
```

```python
hello("Potter", "Mr.") # вывод функции
```

В некоторым случаях мы хотим, чтобы какие-то аргументы можно было не указывать. Скажем, мы
хотим иметь возможность вызвать функцию hello() , определённую выше, не указывая title . В
этом случае сейчас нам выдадут ошибку:


```python
hello("Harry") # вызов функции без параметра title
```

Это неудивительно: мы сказали, что функция hello() должна использовать аргумент title , но не
передали его — какое же значение тогда использовать? Для преодоления этой трудности используютя
значения по умолчанию (default values).


```python
def hello(name, title=""): # объявление функции с заранее определённым значением title
 print("Hello", title, name) # вывод значения 
hello("Harry") # вызов функции
```

```python
hello("Smith", "Mrs.") # вызов функции со всеми параметрами
```

Аргументы можно передавать, указывая их имена.


```python
hello("Smith", title = "Mr.") # вызов функции с именованым параметром
```

```python
hello(name = "Smith", title= "Mr.") # вызов функции с заданными параметрами  
```

В этом случае порядок будет неважен.

```python
hello(title= "Mr.", name = "Smith") # изменение параметра
```

Ещё бывают функции, которые принимают неограниченное число аргументов. Например, так ведёт
себя функция print() .    

```python
print(8, 7, 5, 'hello', 8) # вывод
```

Как она устроена? Примерно вот так:

```python
def my_print(*args): # объявление функции с неизвестным кол-вом параметров
 for x in args: # проходка по списку параметров
    print(x) # вывод значений
```

```python
my_print(6, 8, 9, 'hello', 88, 55) # вызов функции
```

Обратите внимание на звёздочку перед args в сигнатуре функции. Давайте посмотрим
повнимательнее, как работает этот код:

```python
def test(*args): # объявление функции
 print(args) # вывод
test(1,2,3, 'hello') # вызов функции
```

Оказывается, что в args теперь лежит так называемые кортеж, состоящий из элементов, которые

## Отступление: кортежи
Кортеж ( tuple ) — это почти то же самое, что и список, только его элементы неизменяемы.
Обозначается он круглыми скобками.

```python
t = (2, 3, 5, 1) # объявление кортежа
print(t[1]) # вывод первого значения кортежа
print(t[0:2]) # вывод с нулевого по первый значения кортежа включительно
```

```python
for x in t: # проходка по элементам кортежа
 print(x) # вывод
```

```python
t[0] = 10 # изменение нулевого значения кортежа (с ошибкой, кортежи нельзя изменять)
```

```python
t.append(1) # ошибка, нет метода изменения кортежа путём добавления новых элементов
```

Тот факт, что нельзя изменять кортеж, не означает, что нельзя переопределить переменную t :

```python
t = (8, 1, 2, 3) # переопределение переменной
```

```python
t # вывод новой переменной
```

Можно также конвертировать списки в кортежи и наоборот.

```python
print( list( (1, 2, 3) ) ) # вывод перешедшего из кортежа в список объекта 
print( tuple( [1, 2, 3] ) ) # вывод перешедшего из списка в кортеж объекта
```

## Возвращаясь к функциям
Таким образом, звёздочка в сигнатуре как бы ставит дополнительные скобки вокруг аргументов.
Например, следующие два вызова приводят к одинаковым результатам.

```python
def test1(*args): # объявление функции с неограниченным кол-вом параметров
 print(args) # вывод аргументов функции
test1(1, 2, 3) # вызов функции

def test2(args): # объявление функции с одним параметром
 print(args) # вывод параметра
test2( (1,2,3) ) # вызов функции с кортежем в качестве аргумента
```

Можно комбинировать списочные переменные и обычные (при условии, что переменная со
звёздочкой только одна):

```python
def my_print(sep, *args): # объявление функции
 for x in args: # проходка по аргументам
    print(x, end = sep) # вывод значений с окончанием в виде параметра sep
```

```python
my_print('----', 7, 8, 9, 'hello') # вызов функции
```

```python
def connect(list1, list2):
    connectedList = list1 + list2
    print(list1)
    print(list2)
    print(connectedList)
    return connectedList

print(connect([1, 2, 3, 4], ["asd", "dfh", 5]))
```

2. Реализовать функцию перевода значений, веденных
пользователем с клавиатуры. Пользователь вводит значения градусов в
Фаренгейтах, необходимо перевести их в градусы Цельсия.

```python
def toCelsius(fahrenheit  : float) -> float:
    return (fahrenheit - 32) * (5/9)
print(toCelsius(float(input())))
```

3. Реализовать функцию перевода значений, веденных
пользователем с клавиатуры. Пользователь вводит длину в километрах ее
необходимо перевести в мили.

```python
def toMiles(kilometrs : float) -> float:
    return kilometrs / 1.609344
print(toMiles(float(input())))
```

5. Реализовать функцию вывода на экран таблицы умножения числа,
введенного пользователем.

```python
def tabletMult(multiplier : int):
    for i, n in enumerate(range(10)):
        print(i, n*multiplier)
tabletMult(int(input()))
```

6. Реализовать функцию расчета суммы всех двухзначных чисел,
введенных пользователем.

```python
def sumAll(arr : list):
    sum = 0
    for i in arr:
        sum += i
    return sum

arr = []
while True: 
    num = int(input("Введите двухзначное число")) 
    if num >= 10 and num < 100:
        arr.append(num)
    else:
        print(sumAll(arr))
        break
```

7. Реализовать функцию печати всех элементов списка кратных 2.

```python
def divTwoPrint(arr : list):
    for i in arr:
        print(i if i % 2 == 0 else "")
 
arr = []   
while True:
    try:
        num = int(input())
        arr.append(num)
    except:
        divTwoPrint(arr)
        break
```

8. Реализовать функцию печати всех элементов списка кратных 5,
если таких элементов не существует, то необходимо добавить в конец
списка «10».

```python
def divFivePrint(arr : list):
    flag = 0
    for i in arr:
        if i % 5 == 0:
            print(i)
            flag = 1
    if not flag:
        arr.append(10)
    return arr

arr = []   
while True:
    try:
        num = int(input())
        arr.append(num)
    except:
        print(divFivePrint(arr))
        break
```

9. Реализовать функцию печати максимального по модулю элемента
списка не кратного 3.

```python
def maxModNotDivThree(arr):
    max = 0
    for i in arr:
        if i % 3 != 0 and abs(max) < abs(i):
            max = i
    print(i)
    return i

arr = []   
while True:
    try:
        num = int(input())
        arr.append(num)
    except:
        maxModNotDivThree(arr)
        break
```

10. Реализовать функцию расчета произведения четных элементов
списка.

```python
def multDivTwo(arr):
    mult = 1
    for i in arr:
        if i % 2 == 0:
            mult *= i
    return mult

arr = []   
while True:
    try:
        num = int(input())
        print(num)
        arr.append(num)
    except:
        print(multDivTwo(arr))
        break
```
