## Списки и операции с ними

Список — это такая структура данных, которая содержит в себе сразу много элементов.

```python
numbers = [4, 8, 9, 2, 6] # иницилизация списка чисел
```

```python
numbers # вывод
```

В списках можно хранить не только числа. Например, создадим список из строк.

```python
strings = ["Hello", "World", "Test"] # иницилизация списка строк
```

В списке могут храниться данные разных типов. Например, строки, целые числа, числа с плавающей
точкой.

```python
mixed_list = ["Hello", 6, 7.8] # иницилизация смешанного списка
```

Можно обращаться к отдельным элементам списка и работать с ними как с обычными переменными.
Чтобы выбрать элемент нужно указать его номер

```python
print(numbers) # вывод массива чисел
print(numbers[1]) # вывод элемента с индексом 1
```

**Нумерация начинается с нуля!**

> На самом деле, у этого правила есть свои рациональные обоснования https://python-history.blogspot.com/2013/10/why-python-uses-0-based-indexing.html 

Если в списке есть элементы разных типов, они никак не «мешают» друг другу. Например, наличие в
списке строк не превращает другие элементы этого списка в строки.


> Это касается только обычных списков Python. Несколько позже мы будем проходить массивы numpy и там всё не так.

```python
print(mixed_list) # вывод смешанного списка
print(mixed_list[2]+4) # вывод элемента с индексом 2 прибевленного к числу 4
```

Элементы списка можно менять так же, как значения обычных переменных

```python
numbers # вывод списка
```

```python
numbers[1] = 222 # присваивание значения элементу с индексом 1 списка
numbers # вывод списка
```

> Списки Python основаны на стандартных C'шных массивах и обладают их свойствами с точки зрения производительности: в частности, обращение к элементу по его индексу имеет сложность $O(1)$, то есть не является массовой операцией.


Чтобы узнать длину списка, можно использовать функцю len .

```python
len(numbers) # вывод размера списка
```

Заметим, что это не индекс последнего элемента, а именно число элементов. Если вам нужно получить
последний элемент, то его индексом будет len(numbers)-1 . Но в Python можно обращаться к
элементам списка, считая их «с конца», гораздо проще:

```python
numbers = [4, 8, 2, 5] # инициализация списка
numbers[-1] # вывод последнего элемента
```

```python
numbers[-2] # вывод предпоследнего элемента 
```

А вот если вы попытаетесь обратиться к элементу с несуществующим индексом, то получите ошибку.

```python
numbers[5] = 100 # ошибка - выход за размеры списка (попытка присвоить элементу с индексом 5 значения 100)
```

Однако дописывать элементы в конец можно:

```python
numbers = [7, 6, 2] # инициализация списка
print(numbers) # вывод списка
numbers.append(777) # добавление в список значения 777
print(numbers) # вывод списка
```

Слово append — это так называемый «метод» — функция, «принадлежащая» некоторому объекту (в
данном случае — объекту numbers типа list (список)), и что-то делающая с этим объектом. У
numbers , как у любого списка, есть много методов. Можно набрать numbers. , нажать табуляцию
(после точки), и получить список доступных методов. А ещё можно набрать help(list) или даже
help(numbers) (в нашем случае) и получить краткое описание этих методов. Например, так можно
узнать, что помимо append у списков есть метод extend .

```python
print(numbers) # вывод списка
numbers.extend([3, 7, 5]) # составление общего списка из numbers и [3, 7, 5]
print(numbers) # вывод списка
```

Метод extend позволяет приписать к списку сразу несколько элементов. Он получает на вход список,
который нужно приписать: обратите внимание на квадратные скобки внутри круглых при вызове этого
метода — они создают новый список, который и передаётся функции extend . Методы append и
extend меняют список, к которому они применяются. Иногда вместо этого нужно создать новый список,
объединив (конкатенировав!) два других. Это тоже можно сделать.


```python
first_list = [5, 8, 2] # инициализация списка
second_list = [1, 9, 4] # инициализация списка
new_list = first_list + second_list # составление общего списка из first_list и second_list
print(new_list) # инициализация списка
```

Плюсик в данном случае обозначает не поэлементное сложение (как вы могли подумать), а
конкатенацию. Cписки first_list и second_list при этом не изменились


```python
print(first_list) # вывод first_list
print(second_list) # вывод second_list
```

У вас могло возникнуть желание использовать сложение вместо операции extend .


```python
print(numbers) # вывод numbers
# не надо так
numbers = numbers + [2, 6, 9] # много времени из-за копирования
print(numbers) # вывод numbers
```

Вообще говоря, этот код сработал, но делать так не следует: при выполнении операции конкатенации
создаётся новый список, затем в него копируются все элементы из numbers , потом к ним
приписываются элементы из второго списка, после чего старый numbers забывается. Если бы в numbers было много элементов, их копирование в новый список заняло бы много времени. Гораздо
быстрее приписать элементы к уже готовому списку.

## Срезы

Иногда нам нужен не весь список, а его кусочек. Его можно получить, указав в квадратных скобках не
одно число, а два, разделённых двоеточием.


```python
print(numbers) # вывод numbers
print(numbers[1:4]) # вывод numbers с элемента с индексом 1 и до элемента с индексом 4
```

Это называется slice (по-русски часто говорят срез). Обратите внимание: левый конец среза включается
(элемент с индексом 1 — это шестёрка), а правый — нет. Так будет всегда. Это соглашение оказывается
удобным, например, потому что позволяет посчитать число элементов в срезе — нужно из правого конца
вычесть левый (в данном случае 4-1=3).

Если левый элемент не указан, то он считается началом списка, а если правый — то концом

```python
print(numbers[7:]) # с элемента индекса 7
print(numbers[:7]) # до элемента индекса 7
```

Всегда верно следующее: список numbers это то же самое, что numbers[:k]+numbers[k:] , где k —
любой индекс.

Срезы можно использовать для присваивания.

```python
numbers = [5, 8, 9, 10] # инициализация списка
print(numbers[1:3]) # вывод с 1 по 2 элементов массива
numbers[1:3]= [55, 77] # присваивание этим элементам значений 55 и 77
print(numbers) # вывод
```

Не обязательно, чтобы список, который мы присваиваем срезу, имел ту же длину, что и срез. Можно
присвоить более длинный список (тогда исходный список расширится), а можно менее длинный (тогда
сузится). Можно использовать срезы, чтобы вставить несколько элементов внутрь списка. (Для одного
элемента это можно делать с помощью метода insert .)

```python
numbers = [6, 8, 9] # инициализация массива
print(numbers[1:1]) # вывод пустого среза
# это пустой срез
numbers[1:1] = [99, 77, 55] # срез между индексами 1 и 1 получает значение 99, 77 и 55
print(numbers) # вывод
```

> Чтобы вставить какие-то элементы внутрь списка, необходимо освободить для него
место, сдвинув все последующие элементы вперёд. Python сделает это автоматически, но
время это займёт. Поэтому, по возможности, следует этого избегать, особенно если вы
работаете с большими массивами данных. Если вам очень нужно записывать что-нибудь
в начало и конец списка, посмотрите на двустороннюю очередь (deque) из модуля
collections .

Можно удалять элементы списка (и вообще что угодно) или срезы с помощью команды del .


```python
numbers = [6, 7, 9, 12, 8, 3] # инициализация 
del(numbers[4]) # удалим 8
print(numbers) # вывод
del(numbers[0:2]) # удаление элементов с 0 по 1
print(numbers) # вывод
```

Иногда у нас нет никакого конкретного списка, а просто нужно выполнить некоторый код несколько раз,
причём заранее неизвестно, сколько. Для решения этой задачи служит объект range() .

```python
for i in range(5): # 5 циклов 
 print("Hello, i =", i) # вывод "Hello, i =" и индекс цикла
```

range(n) ведёт себя как как список, содержащий целые числа от 0 до n-1 (опять последний элемент
не включается!). В нём ровно n элементов. Ещё можно использовать range в двумя аргументами:
указать начало и конец интервала: range(3,9) . Можно даже превратить range в настоящий список с
помощью команды list :


```python
list(range(5)) # массив чисел с 0 по 5
```

```python
list(range(3,9)) # массив чисел с 3 по 9
```
