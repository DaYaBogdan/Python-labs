## Вещественные числа и погрешности


Вещественные числа в программировании не так просты. Вот, например, посчитаем синус числа π:

```python
from math import pi, sin #импорт из пакета константы Пи и функции синуса
sin(pi) #вычисление синуса Пи
# получится 0? (нет т.к. присутствует шум в переменных и константах типа плавающей точки)
```

Непонятный ответ? Во-первых, это так называемая компьютерная форма экспоненциальной записи
(https://ru.wikipedia.org/wiki/%D0%AD%D0%BA%D1%81%D0%BF%D0%BE%D0%BD%D0%B5%D0%BD%D
чисел. Она удобна, если нужно уметь записывать очень большие или очень маленькие числа: $1.2 \cdot 10^2$
означает, то есть 1200, а $2.4e-3$ — то же самое, что $2.4 \cdot 10^{-3} =  0,0024$. Результат,
посчитанный Python для, имеет порядок — это очень маленькое число, близкое к нулю.
Почему не «настоящий» ноль? Все вычисления в вещественных числах делаются компьютером с
некоторой ограниченной точностью, поэтому зачастую вместо «честных» ответов получаются такие
приближенные. К этому надо быть готовым.

Попробуем теперь написать программу «удвоитель». Она должна будет принимать на вход число,
удваивать его и возвращать результат.


```python
x = input("Введите какое-нибудь число: ") #Ввод и инициализация строки
print(x*2) #вывод 2 строк друг за другом
```

Что-то здесь не то. Конечно, в некотором смысле компьютер сделал то, что мы его просили — было
одно число 12, а стало два, но мы-то хотели чего-то другого. Почему так получилось? Дело в том, что
input всегда возращает строку (он же не знает, что мы введём — цифры или какие-то другие
символы; он не понимает, что в запросе сказан «Введите какое-нибудь число» — для него это просто
какая-то строка). И в переменной x сейчас не число 12 , а строка '12' (обратите внимание на
апострофы).

```python
x #вывод переменной
```

Чтобы работать с ним как с числом, надо его превратить из строкового типа в числовой. Например,
если мы хотим работать только с целыми числами, то подходящим будет тип int (от слова
integer , целое число).

```python
int(x) #переход типа в int
```

Видите, апострофов нет — это уже число. Слово int это одновременно и обозначение типа данных
(целые числа) и функция, которая превращает то, что ей передали в качестве аргумента, в указанный
тип данных.

Перепишем программу-удвоитель следующим образом:

```python
x_str = input("Введите какое-нибудь число: ") #Ввод и инициализация переменной
x = int(x_str) #переход типа в int и инициализация новой переменной
print(x*2) #вывод умноженного x который хранит введённое число на 2
```

Совсем другое дело! Теперь в x_str лежит строчка, которую мы ввели, в x помещается результат
её преобразования в целое число, и дальше это число удваивается и выводится на экран. Можно
было бы не вводить дополнительную переменную (и сэкономить строчку):


```python
x = int(input("Введите какое-нибудь число: ")) #ввод и инициализация числа
print(x*2) #вывод этого числа умноженного на 2
```

Попробуем ещё раз.


```python
x  = int(input("Введите какое-нибудь число: ")) #ввод и инициализация числа
print(x*2) #вывод этого числа умноженного на 2
```

## Практичекое задание: числа Фибоначчи

Числа Фибоначчи или последовательность Фибоначчи — последовательность чисел, начинающаяся
с двух единиц, и такая, что очередное число в этой последовательности равно сумме двух
предыдущих. Формально можно определить её следующим образом:


$a_1 = 1$

$a_2 = 1$

$a_{n+1} = a_n + a_{n−1}$

$n > 2$

Например $a_3 = 1 + 1 = 2; a_4 = 2 + 1 = 3$

**Задача:** посчитать 15-е число Фибоначчи

```python
def Fibonachi(n, An = 1, An_minus_1 = 0):
    return An_minus_1 if n == 0 else (An if n == 1 else Fibonachi(n - 1, An + An_minus_1, An))
    #рекурсивный спуск реализованный чтобы вычислить значение Фибоначчи на нужном нам n
    
print(Fibonachi(int(input("Введите вычесляемое число Фибоначчи: "))))
```

```python
a = 1 # первое число
b = 1 # второе число
i = 2 # номер того числа, которое находится в переменной b (сейчас это a_2)
```

Мы будем хранить два последних числа, поскольку именно они определяют следующее, а также
номер последнего найденного числа (переменная i)

```python
c = a + b # нашли следующее число
i = i + 1 # увеличили i на 1
a = b # значение a нам уже не нужно, а вот значение b ещё пригодится
b = c # запомнили вычисленное значение
print(i, b) #вывод i, а затем b
```

Питон выполняет команды последовательно, строчку за строчкой, поэтому порядок следоавния
команд очень важен. Выполняя эту ячейку несколько раз, вы будете получать каждый раз очередное
число Фибоначчи.
